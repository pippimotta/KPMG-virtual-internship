##Module 2 Data Insight


###RFM Analysis(Recency, Frequency, Monetary)

```python
df_rfm = df_tran_clean.groupby('customer_id').agg({
    'last_purchase':'min',
    'customer_id' : 'count',
     'profit': 'sum'})
df_rfm.rename(columns= {
    'last_purchase': 'recency',
    'customer_id': 'frequency',
    'profit': 'monetary'
}, inplace=True)
df_rfm.head(10)
```

```python
df_rfm.shape
```

```python
quart = df_rfm.quantile(q=[0.25, 0.50, 0.75])
quart
```

```python
#conceptualize/standardize RFM
#for recency, the closer(less) the better
def r_standard(x,p,d):
  if x <= d[p][0.25]:
    return 1
  elif x <= d[p][0.5]:
    return 2
  elif x <= d[p][0.75]:
    return 3
  else: 
    return 4
#for frequency & profit, the more the better
def fm_standard(x,p,d):
   if x <= d[p][0.25]:
    return 4
   elif x <= d[p][0.5]:
    return 2
   elif x <= d[p][0.75]:
    return 3
   else: 
    return 1
```

```python
df_rfm_standard = df_rfm
df_rfm_standard['r_score']= df_rfm.recency.apply(r_standard, args=('recency', quart))
df_rfm_standard['f_score']= df_rfm.frequency.apply(fm_standard, args=('frequency', quart))
df_rfm_standard['m_score']= df_rfm.monetary.apply(fm_standard, args=('monetary', quart))
df_rfm_standard.head()
```

```python
df_rfm_standard['class'] = df_rfm_standard.r_score*100+df_rfm_standard.f_score*10+df_rfm_standard.m_score
df_rfm_standard['total_score'] =df_rfm_standard.r_score+df_rfm_standard.f_score+df_rfm_standard.m_score
df_rfm_standard.head()
```

```python

rfm_quart = (df_rfm_standard.total_score.min(),
             df_rfm_standard.total_score.quantile(q=0.25),
             df_rfm_standard.total_score.median(),
              df_rfm_standard.total_score.quantile(q=0.75),
             df_rfm_standard.total_score.max())
print(rfm_quart)
```

```python
def rfm_class_standard(x, p, d):
  if x <= d[0]:
    return "gold"
  elif x <= d[1]:
    return 'silver'
  elif x <= d[2]:
    return 'bronze'
  else:
    return 'general'
  
df_rfm_standard['classified'] = df_rfm_standard.total_score.apply(rfm_class_standard, args=('total_score', rfm_quart))
df_rfm_standard
```
